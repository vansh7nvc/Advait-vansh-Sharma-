```python
from transformers import pipeline

# 1. Text Classification (Sentiment Analysis)
# Model: distilbert-base-uncased-finetuned-sst-2-english
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
class_output = classifier("I am absolutely loving this new AI tutorial!")

# 2. Summarization
# Model: sshleifer/distilbart-cnn-12-6 (Default)
# The 'summarization' task is not recognized in this environment, using 'text-generation' as a workaround.
summarizer = pipeline("text-generation", model="sshleifer/distilbart-cnn-12-6")
long_text = """
The Hugging Face Transformers library provides thousands of pretrained models to perform
tasks on different modalities such as text, vision, and audio. These models can be used
to compute features, or to perform downstream tasks like classification, summarization,
and translation. It reduces the compute costs and carbon footprint, and saves the time
and resources required to train a model from scratch.
"""
summary_output = summarizer(long_text, max_length=30, min_length=10)

# 3. Question Answering
# Model: distilbert-base-cased-distilled-squad
qa_model = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "The Eiffel Tower is located in Paris, France. It was completed in 1889."
question = "When was the Eiffel Tower completed?"
qa_output = qa_model(question=question, context=context)

# Print Deliverables
print("--- TASK 1: CLASSIFICATION ---")
print(f"Output: {class_output}\n")

print("--- TASK 2: SUMMARIZATION ---")
print(f"Summary: {summary_output[0]['generated_text']}\n")

print("--- TASK 3: QUESTION ANSWERING ---")
print(f"Answer: {qa_output['answer']} (Confidence: {qa_output['score']:.4f})")
```
